```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from matplotlib.dates import DateFormatter

# Load data from CSV files
mois_data = pd.read_csv('mois_data.csv')
tof_rssi = pd.read_csv('tof_rssi.csv')

# Convert 'Time' columns to datetime objects using the specified format
mois_data['Time'] = pd.to_datetime(mois_data['Time'], format='%d:%m:%Y %H:%M:%S')
tof_rssi['Time'] = pd.to_datetime(tof_rssi['Time'], format='%d:%m:%Y %H:%M:%S')
tof_rssi = tof_rssi[tof_rssi['Time of Flight'] >= 8.1]

# Extract relevant columns for Moisture Sensor 1
moisture_sensor_1 = mois_data['Moisture Sensor 1']  # Updated for Moisture Sensor 1
time_mois_data = mois_data['Time']
time_tof_rssi = tof_rssi['Time']
rssi = tof_rssi['RSSI']
tof = tof_rssi['Time of Flight']

# Apply a rolling average to smooth the data
moisture_sensor_1_smoothed = moisture_sensor_1.rolling(window=50, min_periods=1).mean()  # Updated for Moisture Sensor 1
rssi_smoothed = rssi.rolling(window=10, min_periods=0).mean()
tof_smoothed = tof.rolling(window=10, min_periods=0).mean()

# Set figure size
fig, ax1 = plt.subplots(figsize=(16, 3))

# Plot smoothed Moisture Sensor 1 on the left y-axis
ax1.plot(time_mois_data, moisture_sensor_1_smoothed, color='blue', label='Moisture Sensor 1')  # Updated for Moisture Sensor 1
ax1.set_xlabel('Time')
ax1.set_ylabel('Moisture Sensor 1', color='blue')  # Updated for Moisture Sensor 1
ax1.tick_params('y', colors='blue')

# Create a secondary y-axis on the right for smoothed RSSI
ax2 = ax1.twinx()
ax2.plot(time_tof_rssi, rssi_smoothed, color='red', label='RSSI')
ax2.set_ylabel('RSSI', color='red')
ax2.tick_params('y', colors='red')

# Create another secondary y-axis on the right for smoothed Time of Flight
ax3 = ax1.twinx()
ax3.plot(time_tof_rssi, tof_smoothed, color='green', label='Time of Flight')
ax3.set_ylabel('Time of Flight', color='green')
ax3.spines['right'].set_position(('outward', 60))
ax3.tick_params('y', colors='green')

# Set major locator for x-axis to show ticks every 5 days
ax1.xaxis.set_major_locator(MultipleLocator(5))
ax1.xaxis.set_major_formatter(DateFormatter('%d:%m:%Y'))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.title('Moisture Sensor 1, RSSI, and Time of Flight')  # Updated for Moisture Sensor 1
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV files
mois_data = pd.read_csv('mois_data.csv')
tof_rssi = pd.read_csv('tof_rssi.csv')

# Convert 'Time' columns to datetime objects using the specified format
mois_data['Time'] = pd.to_datetime(mois_data['Time'], format='%d:%m:%Y %H:%M:%S')
tof_rssi['Time'] = pd.to_datetime(tof_rssi['Time'], format='%d:%m:%Y %H:%M:%S')
tof_rssi = tof_rssi[tof_rssi['Time of Flight'] >= 8.1]

# Merge dataframes on 'Time'
merged_data = pd.merge(mois_data, tof_rssi, on='Time', how='inner')

# Extract relevant columns for Moisture Sensor 1
vwc = merged_data['Moisture Sensor 1']  # VWC (Moisture Sensor 1)
tof = merged_data['Time of Flight']  # Time of Flight
rssi = merged_data['RSSI']  # RSSI

# Create scatter plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Scatter plot: ToF vs. VWC (ToF on X-axis, VWC on Y-axis)
axes[0].scatter(tof, vwc, color='blue', alpha=0.5)
axes[0].set_xlabel('Time of Flight')
axes[0].set_ylabel('VWC (Moisture Sensor 1)')
axes[0].set_title('ToF vs. VWC')

# Scatter plot: RSSI vs. VWC (RSSI on X-axis, VWC on Y-axis)
axes[1].scatter(rssi, vwc, color='red', alpha=0.5)
axes[1].set_xlabel('RSSI')
axes[1].set_ylabel('VWC (Moisture Sensor 1)')
axes[1].set_title('RSSI vs. VWC')

# Scatter plot: RSSI vs. ToF (RSSI on X-axis, ToF on Y-axis)
axes[2].scatter(rssi, tof, color='green', alpha=0.5)
axes[2].set_xlabel('RSSI')
axes[2].set_ylabel('Time of Flight')
axes[2].set_title('RSSI vs. ToF')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
